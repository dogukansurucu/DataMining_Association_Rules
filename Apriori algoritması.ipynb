## Her ürün için destek değerlerini bulma ve iterasyonlar

```python
# verisetinde her üründen kaç tane olduğuna bakalım.
df.sum()
```

```python
# 1. İterasyon

#Kategorik değişkenlerin destek(support) değerlerini hesaplayarak, bu değerlere göre sıralama yaptık ve buradaki amacımız ürün sıklığını bulmak.
#Tüm öğeleri satır numarasına bölerek destek değerini bulabiliriz.Destek değeri için eşik değeri 0,2 olsun.

# df.sum() ile her bir kategorik değişkenin toplam gözlem sayısını (destek değeri) hesaplar. Bu, her bir kategorik değişkenin kaç kez görüldüğünü temsil eder.
# df.shape[0] ifadesi, veri çerçevesinde bulunan satır sayısını verir. Bu, toplam gözlem sayısını elde etmek için kullanılır.
# Bu iki değeri birbiriyle böler ve böylece her bir kategorik değişkenin destek değerini bulur.

first = pd.DataFrame(df.sum() / df.shape[0], columns = ["Support"]).sort_values("Support", ascending = False)
first
```

```python
# Destek Değerine Göre Eleme

first[first.Support >= 0.20]
```

```python
# 2. İterasyon
# Çift ürün kombinasyonları

# itertools.combinations(first.index, 2) ifadesi, first veri çerçevesindeki indeksler(kategorik değişken adları) arasındaki 
# tüm 2'li kombinasyonları oluşturur. Yani, bu kombinasyonlar iki kategorik değişkenin bir araya getirilmesiyle elde edilir.
# second adlı bir liste oluşturulur ve bu liste, itertools.combinations fonksiyonunun çıktısı olan kombinasyonları içerir.

second = list(itertools.combinations(first.index, 2))
second = [list(i) for i in second]

# Örnek
second[:10]
```

```python
# Çift ürün kombinasyonları için destek değerlerini bulma

# Bir boş liste olan value oluşturulur, bu liste daha sonra her kombinasyonun destek değerlerini içerecektir.
# df.T.loc[second[i]].sum() ifadesi, kombinasyondaki kategorik değişkenlerin her bir gözlem için toplam sayısını verir.
# temp değişkenine bu toplam sayıyı atar.len(temp[temp == df.T.loc[second[i]].shape[0]]) ifadesi, her gözlemdeki toplam sayının,
# kombinasyondaki kategorik değişken sayısına eşit olduğu durumları sayar. Bu sayıyı, tüm gözlemlerin toplam sayısına bölerek destek değerini elde eder.
# Elde edilen destek değeri, value listesine eklenir.

value = []
for i in range(0, len(second)):
    temp = df.T.loc[second[i]].sum() 
    temp = len(temp[temp == df.T.loc[second[i]].shape[0]]) / df.shape[0]
    value.append(temp)
    
    
# Veri çerçevesi oluşturma

# pd.DataFrame kullanılarak, value listesinden bir veri çerçevesi olan secondIteration oluşturulur.
# "index" sütunu, kombinasyonları içeren tuple'lar halinde oluşturulur.
# "length" sütunu, her bir kombinasyondaki kategorik değişken sayısını içerir
# Son olarak, set_index ve sort_values fonksiyonları kullanılarak, destek değerlerine göre sıralama yapılır.

secondIteration = pd.DataFrame(value, columns = ["Support"])
secondIteration["index"] = [tuple(i) for i in second]
secondIteration['length'] = secondIteration['index'].apply(lambda x:len(x))
secondIteration = secondIteration.set_index("index").sort_values("Support", ascending = False)


# secondIteration[secondIteration.Support > 0.1] ifadesi ile destek değeri 0.1'den büyük olan kombinasyonları içeren 
# secondIteration veri çerçevesi filtrelenir.

secondIteration = secondIteration[secondIteration.Support > 0.1]
secondIteration
```

## İterasyonları daha iyi göstermek için bir fonksiyon oluşturuldu.

```python
# Bu fonksiyonda Apriori algoritması türünde birliktelik kuralı analizi yapmayı amaçlar.Fonksiyon, veri setinde belirli bir destek değerini sağlayan 
# kategorik değişken kombinasyonlarını hesaplamak için tasarlanmıştır. İşlemleri şu şekilde açıklayabiliriz:

# ar_iterations fonksiyonu, bir Apriori benzeri birliktelik kuralı analizi gerçekleştirmek üzere tasarlanmıştır.
# Fonksiyonun parametreleri şu şekildedir. "data": Analiz yapılacak veri seti. "num_iter": İterasyon sayısı, yani kaç kategorik değişkenin bir arada inceleneceği.

# support_value: Belirli bir destek değeri eşiği, kombinasyonlar bu eşiği geçen destek değerlerine sahip olmalıdır.
# iterationIndex: İterasyonların oluşturulacağı kategorik değişken kombinasyonlarını belirten bir liste.

# İç içe geçmiş bir fonksiyon olan ar_calculation, verilen kategorik değişken kombinasyonları için destek değerlerini hesaplar 
# ve belirli bir destek değerini sağlamayan kombinasyonları eler.İlk olarak, birinci iterasyon için destek değerleri hesaplanır. 
# Bu, her bir kategorik değişkenin destek değeri, yani gözlemlerde ne kadar sıklıkla bulunduğunu ifade eder.


# Eğer num_iter 1 ise, yani sadece birinci iterasyon için destek değerleri hesaplanıyorsa, bu değerler bir veri çerçevesi olan res içinde saklanır.
# Eğer num_iter 2 ise, yani ikinci iterasyon için kombinasyonlar oluşturulup destek değerleri hesaplanıyorsa, 
# itertools.combinations kullanılarak ikili kombinasyonlar oluşturulur ve ar_calculation fonksiyonu çağrılarak destek değerleri hesaplanır.
# Eğer num_iter 2'den büyükse, yani üçüncü ve sonraki iterasyonlar için kombinasyonlar oluşturulup destek değerleri hesaplanıyorsa, 
# itertools.combinations kullanılarak gerekli kombinasyonlar oluşturulur ve tekrar ar_calculation fonksiyonu çağrılarak destek değerleri hesaplanır.
# Sonuç olarak, hesaplanan destek değerleri ve kombinasyonlar bir veri çerçevesi içinde saklanır ve bu veri çerçevesi döndürülerek birliktelik kuralı analizi sonuçları elde edilmiş olur.
```

```python
def ar_iterations(data, num_iter = 1, support_value = 0.1, iterationIndex = None):
    def ar_calculation(iterationIndex = iterationIndex): 
        value = []
        for i in range(0, len(iterationIndex)):
            result = data.T.loc[iterationIndex[i]].sum() 
            result = len(result[result == data.T.loc[iterationIndex[i]].shape[0]]) / data.shape[0]
            value.append(result)
            
        result = pd.DataFrame(value, columns = ["Support"])
        result["index"] = [tuple(i) for i in iterationIndex]
        result['length'] = result['index'].apply(lambda x:len(x))
        result = result.set_index("index").sort_values("Support", ascending = False)
        result = result[result.Support > support_value]
        return result    
    
    # 1. İterasyon
    first = pd.DataFrame(df.T.sum(axis = 1) / df.shape[0], columns = ["Support"]).sort_values("Support", ascending = False)
    first = first[first.Support > support_value]
    first["length"] = 1
    if num_iter == 1:
        res = first.copy()
        
    # 2. İterasyon
    elif num_iter == 2:
        second = list(itertools.combinations(first.index, 2))
        second = [list(i) for i in second]
        res = ar_calculation(second)     
    else:
        nth = list(itertools.combinations(set(list(itertools.chain(*iterationIndex))), num_iter))
        nth = [list(i) for i in nth]
        res = ar_calculation(nth)
    
    return res
```

```python
# iteration1 değişkeni, ar_iterations fonksiyonunu çağırarak birinci iterasyon sonuçlarını elde eder.
# df: Analiz yapılacak veri seti. num_iter=1: Sadece birinci iterasyon için kombinasyonları inceleyeceğimizi belirtir.
# support_value=0.1: Belirli bir destek değeri eşiği, kombinasyonlar bu eşiği geçen destek değerlerine sahip olmalıdır.
# iteration1 değişkeni, birinci iterasyon sonuçlarını içeren bir veri çerçevesidir. Bu veri çerçevesi, birinci iterasyon için 
# belirlenen destek değeri eşiğini geçen kategorik değişken kombinasyonlarını içerir.
# iteration1 değişkeni ekrana yazdırılarak, birinci iterasyon için bulunan birliktelik kuralı analizi sonuçları gözlemlenir.
# Bu şekilde, iteration1 değişkeni üzerinden birinci iterasyonun sonuçlarına erişilebilir ve analiz sonuçları incelenebilir.

iteration1 = ar_iterations(df, num_iter=1, support_value=0.1)
iteration1
```

```python
iteration2 = ar_iterations(df, num_iter=2, support_value=0.1)
iteration2
```

```python
iteration3 = ar_iterations(df, num_iter=3, support_value=0.01,
              iterationIndex=iteration2.index)
iteration3
```

## Association Kuralları

```python
# Apriori fonksiyonu, veri setinde birliktelik kuralı analizi yapmak için kullanılır. Bu fonksiyon, sıklık tablolarını oluşturarak sık görülen öğe kümelerini belirler. 
# Fonksiyonun parametreleri şunlardır. df: Analiz yapılacak veri seti. min_support=0.1: Belirli bir destek değeri eşiği, bu eşik altında olan öğe kümeleri analiz dışı bırakılır.
# use_colnames=True: Sütun isimlerini kullanarak öğe kümelerini etiketler. verbose=1: Ayrıntılı çıktı almayı sağlar.
# freq_items değişkenine, apriori fonksiyonu çağrılarak elde edilen sık öğe kümeleri (frequent itemsets) atanır.
# freq_items veri çerçevesi, analiz sonuçlarını içerir. Bu veri çerçevesinde her bir öğe kümesinin destek değeri (support) ve öğe kümesini ifade eden sütunlar bulunur.
# freq_items.sort_values("support", ascending=False) ifadesi kullanılarak, destek değerine göre sıralama yapılır. Böylece, en yüksek destek değerine sahip olan öğe kümeleri en üstte yer alır.
# Bu sıralı veri çerçevesi ekrana yazdırılarak, analiz sonuçları görüntülenir. Bu sonuçlar, belirlenen destek değeri eşiğini geçen sık öğe kümelerini içerir. 
# Bu kümeler, birliktelik kuralları çıkarmak ve veri setindeki desenleri anlamak için kullanılabilir.
```

```python
# Apriori
freq_items = apriori(df, min_support = 0.1, use_colnames = True, verbose = 1)
freq_items.sort_values("support", ascending = False)
```

```python
# association_rules fonksiyonu, birliktelik kuralı analizi sonuçları üzerinden kurallar türetmek için kullanılır. Bu fonksiyon, sıklık tabloları ve destek değerleri üzerinden kuralları çıkarmaya yöneliktir. 
# Fonksiyonun parametreleri şunlardır. freq_items: Apriori algoritması ile elde edilen sık öğe kümelerini içeren veri çerçevesi.
# metric="confidence": Kullanılacak ölçüm metriği. Bu durumda, "confidence" yani güvenilirlik kullanılmaktadır.
# min_threshold=0.5: Belirli bir güvenilirlik eşiği, bu eşiğin altındaki kurallar analiz dışı bırakılır.
# df_ar değişkenine, association_rules fonksiyonu çağrılarak kurallar atanır.
# df_ar değişkeni, birliktelik kuralı analizi sonuçları üzerinden türetilen kuralları içeren bir veri çerçevesidir. 
# Bu veri çerçevesinde, her bir kural için destek, güvenilirlik (confidence), lift ve diğer ilgili metrikler bulunur.
# Bu veri çerçevesi ekrana yazdırılarak, birliktelik kuralı analizi sonuçları ve türetilen kurallar görüntülenir. 
# Bu kurallar, sıklıkla birlikte gözlenen öğeler arasındaki ilişkileri temsil eder ve analiz yapılan veri setindeki desenleri anlamak için kullanılır.
```

```python
# Association Kuralları
df_ar = association_rules(freq_items, metric = "confidence", min_threshold = 0.5)
df_ar
```

```python
df_ar[(df_ar.support > 0.15) & (df_ar.confidence > 0.5)].sort_values("confidence", ascending = False)
```
